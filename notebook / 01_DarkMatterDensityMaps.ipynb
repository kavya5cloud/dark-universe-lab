```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

density_map = np.random.normal(
    loc=0,
    scale=1,
    size=(256,256)
)

plt.figure(figsize=(8,8))
plt.imshow(density_map)
plt.colorbar(label="Density")
plt.title("Random Density Field")
plt.show()
```

```python
from scipy.ndimage import gaussian_filter

smooth_map = gaussian_filter(
    density_map,
    sigma=8
)

plt.figure(figsize=(8,8))
plt.imshow(smooth_map)
plt.colorbar(label="Density")
plt.title("Smoothed Density Field")
plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.imshow(
    smooth_map,
    cmap='inferno'
)

plt.colorbar()
plt.title("Synthetic Matter Density")
plt.show()
```

```python
fft_map = np.fft.fft2(smooth_map)

print("Shape:", fft_map.shape)
print("Datatype:", fft_map.dtype)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(smooth_map, cmap='inferno')
plt.colorbar()

plt.savefig(
    "density_map.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
